```python
cipherCodeList = []
increment = 0

# this is the code which generates our string A - Z
for i in range(65,91):
    cipherCodeList.insert(increment,chr(i))       # this converts the 65 number into its ASCI equivalent which is A, and then inserts it into the list
    increment += 1
    
text = input("Enter the key:")

# this (the next 3 lines ) takes our key and coverts it into a list and makes it into upper case and prints it !!
a = []
a = list(text.upper())
print("This is the entered key converted to string")
print(a)

lenOfKey = len(a)


# this removes the common elements from the list
cipherCodeList2 = list(set(a)^set(cipherCodeList))
cipherCodeList2.sort()

# This inserts our key into the cipher code
for i in range (len(a)):
    cipherCodeList2.insert(i,a[i])

print("The cipherCodeList after removing the similar words will be\n")
print(cipherCodeList2)

# this deletes any element which has an index more than 25    
del cipherCodeList2[25:]


lenOfcipherCodeList = len(cipherCodeList2)

# indexOfItemsToDelete = lenOfcipherCodeList - lenOfKey

print("This is the length of the cipher code text\n"+str(lenOfcipherCodeList))
print("This the cipher code text")
print(cipherCodeList2)


# the below code creates 5 by 5 matrix of the cipherCodeList2 list

matrix = []
stringlen1 = len(cipherCodeList2)
stringlen1 = stringlen1/5
a =0
b =5
inc1 = 0
for i in range(int(stringlen1)):
    matrix.insert(inc1,cipherCodeList2[a:b])
    a = b
    b = b + 5
    inc1 +=1
    
print("This is the matrix of the cipher code text")
print(matrix)

wordToBeCiphered = (input("Enter the word that you need to create a cipher code of !!")).upper()
print("The word that we need to create a cipher text is")
print(wordToBeCiphered)


# this is the code to create a pair of words of a string and store each pair in a list one by one

pairOfKeyWords = []
stringlen = len(wordToBeCiphered)
stringlen = stringlen/2
a =0
b =2
inc = 0
for i in range(int(stringlen)):
    pairOfKeyWords.insert(inc,wordToBeCiphered[a:b])
    a = b
    b = b + 2
    inc +=1
    
print(pairOfKeyWords)


for i in pairOfKeyWords:
    print(i)
    

# this creates a list of list    
pairOfKeyWords = [list(i) for i in pairOfKeyWords]
print(pairOfKeyWords)

# this is how we can iterate over a list inside a list 
# if it shows that array index is out of range then that means, the index that we are putting does not have any element assigned to it 
# for ex: you have 3 elements in an array and you are putting index value 4, but for index value 4 there is no element
print("this print any element we want from the list of the list")
print(pairOfKeyWords[1][1])

## NOW WE NEED TO ADD CONDITIONS OF THE MATRIX


```
